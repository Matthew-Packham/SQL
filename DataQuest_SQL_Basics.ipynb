### GROUP BY

```python
#we can do the above queries in one using GROUP BY!!!

##GROUP BY separates rows into groups based on the contents 
##of a particular column and allows us to perform aggregate functions on each group


query = """
SELECT sub_type, AVG(duration) AS 'Average Duration'
FROM trips
GROUP BY sub_type"""

run_query(query, conn)

#amazing!!!
```

```python
### question: which bike was used for the most trips?

query="""
SELECT bike_number AS 'Bike Num', COUNT(bike_number) AS 'Bike frequency'
FROM trips
GROUP BY bike_number 
ORDER BY COUNT(bike_number) DESC
LIMIT 3"""

run_query(query, conn)

## We use GROUP BY to tell the COUNT function on line 1 to count up each value for bike_number separately.
```

### JOIN !!!

So far we’ve been looking at queries that only pull data from the trips table. However, one of the reasons SQL is so powerful is that it allows us to pull data from **multiple** tables in the same query. 

Our bike-sharing database contains a second table, stations. The stations table contains information about every station in the Hubway network and includes an id column that is referenced by the trips table.

#### Basics

* To tell sql which table a column is from we use a dot (like a method in Python). i.e. SELECT trips.duration (the duration column is in the trips table)

* To tell the database how the trips and stations tables are connected, we use JOIN and ON. JOIN specifies which tables should be connected and ON specifies which columns in each table are related.

We’re going to use an inner join, which means that rows will only be returned where there is a match in the columns specified in ON. In this example, we will want to use JOIN on whichever table we didn’t include in the FROM command. So we can either use FROM trips INNER JOIN stations or FROM stations INNER JOIN trips.

As we discussed earlier, these tables are connected on the customer_id column in each table. Therefore, we will want to use ON to tell the database that these two columns refer to the same thing like this:

ON orders.customer_ID = customers.customer_id

IMPORTANT NOTE: for INNER JOIN (aka JOIN) it displays only attributes that match in each tables (based on the ON clause) i.e. if a order in orders.customer_ID is not presnet in custermors.customer_id - we will lose that entry!

#### Question: Which station is the most frequent starting point?

* First we want to use SELECT to return the station column from the stations table and the COUNT of the number of rows.
* Next we specify the tables we want to JOIN and tell the database to connect them ON the start_station column in the trips table and the id column in the stations table.
* Then we get into the meat of our query – we GROUP BY the station column in the stations table so that our COUNT will count up the number of trips for each station separately
* Finally we can ORDER BY our COUNT and LIMIT the output to a manageable number of results

```python
query = """
SELECT stations.station AS 'Station', COUNT(*) AS 'Count'
FROM trips INNER JOIN stations
ON trips.start_station = stations.id
GROUP BY stations.station 
ORDER BY COUNT(*) DESC
LIMIT 5;
"""
run_query(query, conn)

### key bit is that we group by the stations and tell sql to count how many times that station comes up
### but since our infomation about the station names is found in stations table we have to get that info
### join the station id to the id in start station!
```

#### Question: Which stations are most frequently used for round trips?

We can use much the same query as before. We will SELECT the same output columns and JOIN the tables in the same way, but this time we’ll add a WHERE clause to restrict our COUNT to trips where the start_station is the same as the end_station.

```python
query ="""
SELECT stations.station AS 'Station', COUNT(*) AS 'Count'
FROM trips INNER JOIN stations
ON trips.start_station = stations.id
WHERE trips.start_station = trips.end_station
GROUP BY stations.station
ORDER BY COUNT(*) DESC
LIMIT 5;"""

run_query(query, conn)
```

#### Question: How many trips start and end in different municipalities?

municipality is area in US! 

We want to know how many trips start and end in a different municipality. To achieve this, we need to JOIN the trips table to the stations table twice. Once ON the start_station column and then ON the end_station column.

In order to do this, we have to create an alias for the stations table so that we are able to differentiate between data that relates to the start_station and data that relates to the end_station. We can do this in exactly the same way we’ve been creating aliases for individual columns to make them display with a more intuitive name, using AS.

For example we can use the following code to JOIN the stations table to the trips table using an alias of ‘start’. We can then combine ‘start’ with our column names using . to refer to data that comes from this specific JOIN (rather than the second JOIN we will do ON the end_station column):

```python
query = """
SELECT COUNT(trips.id) AS 'Number of Trips staring and ending in different Municipals'
FROM trips INNER JOIN stations AS 'Start'
ON trips.start_station = Start.id
INNER JOIN stations as 'End'
ON trips.end_station = End.id
WHERE Start.municipality <> End.municipality """
##<> means not equal to
run_query(query, conn)

### key pt: we need to do a double join! we need 1st JOIN between trips and stations (based on trips.start_station = start.id)
### (start.id is the id of the stations) which we call start -- we then create 
### a further join between start and stations again (but this time based on trips.end_station = End.id) and call this end. 

### now we have a way of relating the the start_station to its municipal (in the stations table) and like wise the end_station
### to its municipal. 
### we can therefore call the start join and look for the municipal and likweise we can call the end join and look at its
### municipal. and where they are different - we count it!
```
